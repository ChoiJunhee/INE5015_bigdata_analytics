```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
KBS_MMS_All =  pd.read_csv('./[0]_KBS_MMS_All.csv')
KBS_STD_All =  pd.read_csv('./[1]_KBS_STD_All.csv')
RFE_MMS_All =  pd.read_csv('./[2]_RFE_MMS_All.csv')
RFE_STD_All =  pd.read_csv('./[3]_RFE_STD_All.csv')
```

```python
#하나씩 확인

import matplotlib.pyplot as plt

import seaborn as sns

plt.rcParams['figure.figsize'] = [10, 6]



sns.boxplot(x="Pass/Fail", 

            y="F27", 

            hue="Pass/Fail",

            data=RFE_STD_All)

plt.show()
```

```python
features = list(RFE_STD_All.columns)

for feature in features:
    sns.boxplot(x="Pass/Fail", 

                y=RFE_STD_All[feature], 

                hue="Pass/Fail",

                data=RFE_STD_All)

    plt.show()
```

```python
features1 = list(RFE_MMS_All.columns)

for feature in features1:
    sns.boxplot(x="Pass/Fail", 

                y=RFE_MMS_All[feature], 

                hue="Pass/Fail",

                data=RFE_MMS_All)

    plt.show()
```

```python
features2 = list(KBS_STD_All.columns)

for feature in features2:
    sns.boxplot(x="Pass/Fail", 

                y=KBS_STD_All[feature], 

                hue="Pass/Fail",

                data=KBS_STD_All)

    plt.show()
```
