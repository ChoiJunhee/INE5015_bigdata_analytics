```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
KBS_MMS_All =  pd.read_csv('./[0]_KBS_MMS_All.csv')
KBS_STD_All =  pd.read_csv('./[1]_KBS_STD_All.csv')
RFE_MMS_All =  pd.read_csv('./[2]_RFE_MMS_All.csv')
RFE_STD_All =  pd.read_csv('./[3]_RFE_STD_All.csv')

RFE_STD_All = RFE_STD_All.drop(['Time'], axis=1)
RFE_MMS_All = RFE_MMS_All.drop(['Time'], axis=1)
KBS_STD_All = KBS_STD_All.drop(['Time'], axis=1)
KBS_MMS_All = KBS_MMS_All.drop(['Time'], axis=1)
```

```python
features = list(RFE_STD_All.columns)



plt.figure( figsize=(40,250))
for i, feature in enumerate(features) :
    #print(feature)
    plt.subplot(30 ,4 ,i+1)
    sns.boxplot(x="Pass/Fail", 

                y=RFE_STD_All[feature], 

                #hue="Pass/Fail",
                
                data=RFE_STD_All)
    

plt.show()
```
