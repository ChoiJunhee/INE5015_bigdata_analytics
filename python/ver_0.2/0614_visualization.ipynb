```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


RFE_MMS = pd.read_csv('./RFE_MMS.csv')
RFE_MMS = RFE_MMS.drop(['Time'], axis=1)


features = list(RFE_MMS.columns)

plt.figure( figsize=(40,250))
for i, feature in enumerate(features) :
    #print(feature)
    plt.subplot(33 ,4 ,i+1)
    sns.boxplot(x="Pass/Fail", 

                y=RFE_MMS[feature], 

                #hue="Pass/Fail",
                
                data=RFE_MMS)
    

plt.show()
```
